```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from scipy import signal
```

```python
# generate a square wave from numpy only
def generate_square_wave(amplitude, frequency, samples_per_second, duration):
    time_vector = np.arange(0, duration, 1/samples_per_second)  
    wave = amplitude * np.sign(np.sin(2 * np.pi * frequency * time_vector))  
    return time_vector, wave
```

```python
# generate a square wave using the signal module from scipy
def generate_square_wave2(amplitude, frequency, samples_per_second, duration):
    "””Use numpy to generate the time vector and signal module from scipy to generate the wave””"
    time_vector = np.arange(0, duration, 1/samples_per_second)
    wave = amplitude * signal.square(2 * np.pi * frequency * time_vector)
     # wave is a numpy array of shape (samples_per_second* during, 1) so for this case it has shape (400,1)
    return time_vector, wave
```

```python
# plot using matplotlib

def plot_and_save_wave(t, wave, csv_file="square_wave.csv", wave_file="square_wave.png" ):
    plt.plot(t, wave)
    plt.title('Square Wave')
    plt.xlabel('Time')
    plt.ylabel('Amplitude')
    plt.grid(True)
    plt.savefig(wave_file)
    plt.show()

    df = pd.DataFrame({'Time': t, 'Amplitude': wave})
    print(df.head())
    df.to_csv(csv_file, index = False)
    print(f'Waveform saved t o {csv_file} and wave diagram saved to {wave_file}')
```

```python
# plot using seaborn and matplotlib
def plot_and_save_wave2(time_vector, wave, csv_file="square_wave.csv", wave_file="square_wave.png" ):

    """ Use pandas to convert the time vector, and wave into a pandas dataframe,  and  Plot the square wave using seaborn and matplotlib, save the dataframe to csv_file and the plot to wave_file"""

    df = pd.DataFrame({'Time': time_vector, 'Amplitude': wave})

    sns.set_theme(style="darkgrid")  
    plt.figure(figsize=(10, 6)) 
    sns.lineplot(x='Time', y='Amplitude', data=df)
    plt.title('Square Wave')
    plt.xlabel('Time (s)')
    plt.ylabel('Amplitude')
    plt.savefig(wave_file)
    plt.show()

    print(df.head()) # print first 5 rows that is first five samples, for first x samples use df.head(x)
    df.to_csv(csv_file, index = False) # save the dataframe to a csv file
    print(f'Waveform saved to {csv_file} and wave diagram saved to {wave_file}')
```

```python
amplitude = 5 # can take input using input()
frequency = 1
samples_per_second = 100
duration = 4
```

```python
t, wave = generate_square_wave(amplitude, frequency, samples_per_second, duration)
plot_and_save_wave(t, wave)
```
